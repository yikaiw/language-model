```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
train_perp = []
valid_perp = []
with open('nohup.out', 'r') as f:
    for line_ in f.readlines():
        line = line_.split()
        if 'progress' not in line:
            if 'Train' in line and 'perplexity' in line:
                train_perp.append(float(line[-1]))
            elif 'Valid' in line and 'perplexity' in line:
                valid_perp.append(float(line[-1]))
```

```python
fig, ax = plt.subplots(figsize=(13, 6))
x = np.arange(len(train_perp[12:]))
ax.plot(x, train_perp[12:], '', label='train perplecity')
ax.plot(x, valid_perp[12:], '', label='valid perplecity')
ax.set_xlabel('epoch', fontsize=13)
plt.xticks(np.arange(0, len(train_perp[12:]), 2))
plt.yticks(np.arange(0, 450, 50))
plt.grid(True, linestyle = '-.')  
legend = ax.legend(loc='upper right', fontsize='x-large')
```
